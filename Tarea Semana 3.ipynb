# Apliaci√≥n de carga axial de 100 kpi

**Punto 4**

```python
# C√≥digo de almacenamiento de datos:

mats = ['Acero 1020','Acero 1040','Aluminio 7075','Lat√≥n','Tit√°neo','Nylon','GFRP']
E = [210,210,70,130,100,3,1.56]  # GPa
rho = [7.8,7.8,2.7,8.4,4.5,1.1,1.8] # Mg/m3
k = [0.25, 0.5, 1.9, 2.2, 16.25, 4.3, 110] # $/kg
```

* a) minimizar el di√°metro 

Para minimizar el di√°metro se tiene en cuenta la siguiente expresi√≥n:

\begin{equation}
    \frac{4F}{\pi d^2} = \frac{E}{\epsilon}
\end{equation}

Resolviendo para el di√°metro, tenemos una simplificaci√≥n a causa de sacar a un lado las constantes del problema, como la carga, œÄ, y la deformaci√≥n, cuya multiplicaci√≥n se define como $\kappa$:

\begin{equation}
    d = \kappa \sqrt{ \frac{1}{E}}
\end{equation}

Esto se traduce en que hay que encontrar el material con mayor M√≥dulo de Elasticidad  


```python
print('El material con mayor m√≥dulo de elasticidad, el cual disminuye el di√°metro es:',mats[E.index(max(E))]  )
```

* b) Minimizar el peso

Para esto se escoge el de menor relaci√≥n $\rho^2 / E$, entonces tenemos que:

```python
from operator import mul, truediv
multi = list(map(mul,rho,rho))
lista_final = list(map(truediv,multi,rho))

print('Para minimizar el peso se tiene que escoger',mats[lista_final.index(min(lista_final))])
```

* c) Minimizar el costo

Para minimizar el costo, tenemos que encontrar el valor que tenga la peor raz√≥n de $k  \rho$, ya que este valor es proporcional al costo total del material. 

```python
from operator import mul
rela = list(map(mul,rho,k))
print('El material m√°s barato es',mats[rela.index(min(rela))])
```

* d) Minimizar la enlongaci√≥n axial:

Para esto, tenemos que referirnos a la expresion enunciada en el literal a), en la que resolviendo para la deformaci√≥n unitaria se tiene que:

\begin{equation}
    \epsilon = \frac{E}{\rho F L}
\end{equation}

Sacando las constantes deducimos que se tiene que escoger el material con menor relaci√≥n $E / \rho$

```python
from operator import truediv
rel = list(map(truediv,E,rho))
print('El material que hay que escoger es',mats[rel.index(min(rel))])
```
